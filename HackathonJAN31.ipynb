```python
import pandas as pd
import matplotlib.pyplot as mpl
```

```python
orders_data=pd.read_csv("orders.csv")
```

```python
orders_data.info()
```

```python
orders_data.columns
```

```python
orders_data['order_date']=pd.to_datetime(orders_data['order_date'],dayfirst=True)
```

```python
user_data=pd.read_json("users.json")
```

```python
orders_data
```

```python
import sqlite3

connection=sqlite3.connect("restaurants.db")
restaurants_data = pd.read_sql("SELECT * FROM restaurants", connection)
```

```python
restaurants_data
```

```python
orders_data
```

```python
user_data
```

```python
final_data=pd.merge(orders_data,restaurants_data,on="restaurant_id",how="left")
```

```python
final_data
```

```python
final_data=pd.merge(final_data,user_data,on="user_id",how="left")
```

```python
final_data
```

```python
final_data.to_csv("final_food_delivery_dataset.csv",index=False)
```

### Which city has the highest total revenue (total_amount) from Gold members?  

```python
final_data[final_data['membership']=="Gold"].groupby("city")["total_amount"].sum().sort_values(ascending=False)
```

### Which cuisine has the highest average order value across all orders?

```python
final_data.groupby("cuisine")["total_amount"].mean().sort_values(ascending=False)
```

### How many distinct users placed orders worth more than ₹1000 in total (sum of all their orders)?

```python
distint_users = (final_data.groupby("user_id")["total_amount"].sum())
```

```python
count=0
for x in distint_users:
    if(x>1000):
        count+=1;
```

```python
count
```

### Which restaurant rating range generated the highest total revenue?

```python
final_data.groupby("rating")["total_amount"].mean().sort_values(ascending=False)
```

### Among Gold members, which city has the highest average order value?

```python
final_data[final_data['membership']=="Gold"].groupby("city")["total_amount"].mean().sort_values(ascending=False)
```

### Which cuisine has the lowest number of distinct restaurants but still contributes significant revenue?

```python
pd.pivot_table(final_data,index="cuisine",values=["restaurant_id", "total_amount"],aggfunc={"restaurant_id": pd.Series.nunique,"total_amount": "sum"})
```

### What percentage of total orders were placed by Gold members? (Rounded to nearest integer)

```python
final_data.groupby("membership")["total_amount"].sum().sort_values(ascending=False).iloc[1]/(final_data['total_amount'].sum())
```

### Which restaurant has the highest average order value but less than 20 total orders?

```python
order_value=final_data.groupby("restaurant_name_x") .agg(total_orders=("order_id", "count"), avg=("total_amount", "mean"))
```

```python
order_value.loc[["Grand Cafe Punjabi","Grand Restaurant South Indian","Ruchi Mess Multicuisine","Ruchi Foods Chinese"]]
```

### Which combination contributes the highest revenue?

```python
final_data.groupby(["membership", "cuisine"])["total_amount"] .sum() .sort_values(ascending=False)
```

### During which quarter of the year is the total revenue highest?

```python
final_data.groupby(final_data["order_date"].dt.to_period("Q"))["total_amount"].sum().idxmax()
```

### How many total orders were placed by users with Gold membership?

```python
final_data.groupby("membership")["total_amount"].sum()
```

### What is the total revenue (rounded to nearest integer) generated from orders placed in Hyderabad city?

```python
final_data.groupby("city")["total_amount"].sum().sort_values(ascending=False)
```
