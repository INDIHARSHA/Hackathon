```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd




final_data=pd.read_csv("final_food_delivery_dataset.csv")
final_data['order_date']=pd.to_datetime(final_data['order_date'],dayfirst=True)
quarterly_revenue = final_data.groupby(final_data["order_date"].dt.to_period("Q"))["total_amount"].sum()

sns.barplot(x=quarterly_revenue.index.astype(str), y=quarterly_revenue.values, palette="viridis")
plt.title("Total Revenue by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.show()
```

```python
top_restaurants = final_data.groupby("restaurant_name_x")["total_amount"].sum().nlargest(10)

top_restaurants.plot(kind="bar", figsize=(10,5), color="skyblue")
plt.title("Top 10 Restaurants by Revenue")
plt.xlabel("Restaurant")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
membership_counts = final_data["membership"].value_counts()

membership_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(6,6))
plt.title("Orders by Membership Type")
plt.ylabel("")
plt.show()
```
